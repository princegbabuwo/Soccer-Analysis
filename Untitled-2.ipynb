### Psedocode
1. Filter for Full Time fixtures:

```python
#read json file

with open(json_file) as file:
        data = json.load(file)
```

```python
fixtures = data['response']
fixtures
```

```python
fixturesDF = pd.json_normalize(fixtures, sep='.')
fixturesDF
```

```python
fixturesDF.iloc[0]
```

```python
def extract_bookmakers(fixturesDF):
    bookmakers_list = []
    for _, row in fixturesDF.iterrows():
        fixture_id = row['fixture.id']
        bookmakers = row['bookmakers']

        # For each bookmaker, extract the relevant details
        for bookmaker in bookmakers:
            bookmaker_id = bookmaker['id']
            bookmaker_name = bookmaker['name']
            if bookmaker_name.lower() != "1xBet".lower():
                 continue
            bets = bookmaker['bets']


            #TODO: Check for total markets ("name": "Goals Over/Under") for over 0.5 and under 0.5 only
            # For each bet within a bookmaker, create a new entry
            for bet in bets:
                bet_id = bet['id']
                bet_name = bet['name']
                # Append data as a new row in the list
                bookmakers_list.append({
                    'fixture.id': fixture_id,
                    'bookmaker.id': bookmaker_id,
                    'bookmaker.name': bookmaker_name,
                    'bet.id': bet_id,
                    'bet.name': bet_name
                })

    bookmakersDF = pd.DataFrame(bookmakers_list)
    return bookmakersDF
```

```python
bookmakersDF = extract_bookmakers(fixturesDF)
bookmakersDF
```

```python
# Lets filter for FullTime Matches only

ftFixturesDF = fixturesDF[fixturesDF['fixture.status.short'] == 'FT']
ftFixturesDF
```
